# SAFT-$\gamma$ Mie using NET-GP

## Requirement

The following python packages are required:
- [sgtpy](https://github.com/gustavochm/sgtpy)
- numpy

## Methodology

Non-equilibrium polymer phase $(NE)$:
1. Polymer phase density $(\rho_{2})$ is known a priori.
2. Chemical potential of the solvent in the polymer phase $(\mu_{1}^{NE})$ is a function of $(T,x_{1},\rho_{2})$ , with only unknown is $x_{1}$. $\mu_{1}^{NE}(T,x_{1},\rho_{2})$ can be calculated calculated directly with sgtpy package.

Equilibrium exeternal gas phase $(EQ)$:
1. Assuming gas phase is pure solvent.
2. Chemical potential of solvent in the external gas phase $(\mu_{1}^{EQ})$ is a function of $(T,P)$. This can be calculated with any equilibrium Equation of States (e.g. SL EOS, SAFT etc.).

Phase equilibrium:
1. Condition:<br>$\mu_{1}^{NE}(T,x_{1},\rho_{2}) = \mu_{1}^{EQ}(T,P)$
2. The equation is solved for composition $x_{1}$, hence solubility is obtained.

## Sensitivity Analysis

### Sensitivity on $MW_{2}$

```python
plot_MW2_sensitivity_EQvNE(MW2_l=250e3,MW2_u=300e3,no_of_points=20,T=temp, p=pressure,sol=solute, pol=polymer)
```

*On the role of $MW_{2}$*:

$MW_{2}$ is used at different stage in NE and EQ version of the EOS:
1. **EQ**: Firstly, molar solubility is solved based on $\mu$ of both phaes calculated  from T and P. Then, $MW_{2}$ is used to convert that molar solubility to mass solubility.
2. **NE**: $MW_{2}$ is used together with mass density of the glassy polymer to calculate the molar density of glassy polymer phase, which is an independent variable to calculate $\mu$ of glassy polymer phase gas-polymer mixture.

## Comparison between NE and EQ

### Isotherm

3 curves overall are plotted:
1. **NE** with $k_{sw}=0$
2. **NE** with $k_{sw}$ value specified in function argument. If not provided, base value from database with be used.
3. **EQ** curve version of EOS.

In all cases, there is an option to manually specify $MW_{2}$ value. This will be used for calculations and plotting instead of $MW_{2}$ base value from database.

```python
plot_isotherm_EQvsNE(
        p_l=0.001e6,    # lower bound
        p_u=10e6,       # upper bound
        no_of_points=20,# no of points inbetween bounds
        T=temp,
        sol=solute,
        pol=polymer,
        k_sw=0.00559284,        # manually specified value
        MW_2=267820,        # manually specified value
    )
```

*Note that EQ line may not be fully solved. This can be fixed by trying different x0 values in *solve_solubility_EQ* function in test_sgtpy.py.*
